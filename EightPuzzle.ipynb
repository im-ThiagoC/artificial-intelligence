Solução para o problema EightPuzzle

Função principal onde é definido o estado inical

```python
import copy
#Estrutura do EightPuzzle

class eightPuzzle:
  def __init__(self, puzzle):
    self.puzzle = puzzle
    self.size = len(puzzle)
    self.chegada = list(range(1, self.size)) + [0]
    print("Chegada: ", self.chegada)

  def getPeçaVazia(self, estado):
    return estado.index(0)

#Função onde a peça será movida
  def moverPeça(self, direction):
    peçaVazia = self.puzzle.index(0)
    linhaVazia = peçaVazia // self.size
    colunaVazia = peçaVazia % self.size

    if direction == "up" and linhaVazia > 0:
      novaLinha = linhaVazia - 1
      novoIndex = novaLinha * self.size + colunaVazia
      self.puzzle[peçaVazia] = self.puzzle[novoIndex]
      self.puzzle[novoIndex] = self.puzzle[peçaVazia]

    elif direction == "down" and linhaVazia < (self.size - 1):
      novaLinha = linhaVazia + 1
      novoIndex = novaLinha * self.size + colunaVazia
      self.puzzle[peçaVazia] = self.puzzle[novoIndex]
      self.puzzle[novoIndex] = self.puzzle[peçaVazia]

    elif direction == "left" and colunaVazia > 0:
      novaColuna = colunaVazia - 1
      novoIndex = linhaVazia * self.size + novaColuna
      self.puzzle[peçaVazia] = self.puzzle[novoIndex]
      self.puzzle[novoIndex] = self.puzzle[peçaVazia]

    elif direction == "right" and colunaVazia < (self.size - 1):
      novaColuna = colunaVazia + 1
      novoIndex = linhaVazia * self.size + novaColuna
      self.puzzle[peçaVazia] = self.puzzle[novoIndex]
      self.puzzle[novoIndex] = self.puzzle[peçaVazia]

  #Verifica se é a chegada
  def is_chegada(self):
    return self.puzzle == self.chegada

  #Retorna os movimentos possíveis
  def movimentosLivres(self):
    peçaVazia = self.getPeçaVazia(puzzle)
    linhaVazia = peçaVazia // self.size
    colunaVazia = peçaVazia % self.size

    moves = []
    if peçaVazia >= 3:
      moves.append("up")
    if peçaVazia < 6:
      moves.append("down")
    if peçaVazia % 3 != 0:
      moves.append("left")
    if (peçaVazia + 1) % 3 != 0:
      moves.append("right")

    return moves

def eightPuzzleSolver(initial_state):
  inicio = [(initial_state, [])]
  explorados = set()

  while inicio:
    estado, caminho = inicio.pop(0)
    explorados.add(tuple(estado.puzzle))

    if estado.is_chegada():
      return caminho

    for move in estado.movimentosLivres():
      novoEstado = copy.deepcopy(estado)
      novoEstado.moverPeça(move)
      if tuple(novoEstado.puzzle) not in explorados:
        inicio.append((novoEstado, caminho + [move]))

  return None
```

```python
import random

if __name__ == "__main__":
  puzzle = [1, 2, 3, 4, 5, 6 ,7 ,8 ,0]
  random.shuffle(puzzle)
  puzzle.append(0)
  print("Estado inicial:", puzzle)
  initial_state = eightPuzzle(puzzle)
  print(initial_state)
  solution = eightPuzzleSolver(initial_state)

  if solution:
    print("Solução encontrada em ", solution, "passos:")
    for step, move in enumerate(solution, 1):
      print("Passo ", step, ":", move)
      initial_state.moverPeça(move)
      print(initial_state)
  else:
    print("Sem solução.")
```
